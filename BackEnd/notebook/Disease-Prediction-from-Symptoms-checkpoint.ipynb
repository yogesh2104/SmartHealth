## Disease Prediction from Symptoms


```python
# Import Dependencies
import csv
import pandas as pd
import numpy as np
from collections import defaultdict
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the file
df_concat = pd.read_csv('./dataset/dis_sym_dataset_comb.csv')
df_concat.head()
```

```python
df_concat.drop_duplicates(keep='first',inplace=True)
```

```python
df_concat.head()
```

```python
df_concat.describe()
```

```python
len(df_concat)
```

```python
cols = df_concat.columns
cols
```

```python
cols = cols[1:]
```

```python
# Since, every disease has multiple symptoms, combine all symptoms per disease per row
df_concat = df_concat.groupby('disease').sum()
df_concat = df_concat.reset_index()
df_concat[:5]
```

```python
len(df_concat)
```

```python
# df_concat.to_csv("./dataset/training_dataset.csv", index=False)
```

```python
# One Hot Encoded Features
X = df_concat[cols]

# Labels
y = df_concat['disease']
print(X)
print(y)
```

## Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
len(X_train), len(y_train)
```

```python
len(X_test), len(y_test)
```

```python
dt = DecisionTreeClassifier()
clf_dt=dt.fit(X, y)
```

```python
clf_dt.score(X, y)
```

```python
export_graphviz(dt, 
                out_file='./tree.dot', 
                feature_names=cols)
```

```python
disease_pred = clf_dt.predict(X)
```

```python
disease_real = y.values
```

```python
for i in range(0, len(disease_real)):
    if disease_pred[i]!=disease_real[i]:
        print ('Pred: {0}\nActual: {1}\n'.format(disease_pred[i], disease_real[i]))
```
